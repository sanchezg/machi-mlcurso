Carga del dataset original.

```python
from kickstart import load_data

dataset = load_data("digits_train.jsonlines")
```

Crear una lista que me va a contener una media de la gray scale para cada pixel de cada digito en el dataset original.

```python
digits_mean = [[0]*784]*10

for s in dataset:
    digits_mean[int(s['label'])] = [sum(x) for x in zip(digits_mean[int(s['label'])], s['image'])]
```

```python
from collections import Counter

c = Counter(s['label'] for s in dataset)
```

```python
from operator import div

for idx in xrange(len(digits_mean)):
    digits_mean[idx] = map(div, digits_mean[idx], [c[str(idx)]]*784)
```

```python
from itertools import islice

def draw_ascii_image(image):
    it = iter(image)
    for _ in range(28):
        line = []
        for b in islice(it, 28):
            if b < 80:
                c = "."
            elif b < 160:
                c = "*"
            else:
                c = "#"
            line.append(c * 2)
        print("".join(line))
```

```python
for idx in xrange(len(digits_mean)):
    draw_ascii_image(digits_mean[idx])
```

Ahora divido el dataset original en train y test sets.

```python
from random import shuffle
shuffle(dataset)

ltrain = int(len(dataset) * 0.9)

train = dataset[:ltrain]
test = dataset[ltrain:]

print 'Dataset length: ', len(dataset)
print 'Train length: ', len(train)
print 'Test length: ', len(test)
```
